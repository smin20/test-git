```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 데이터 생성
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 3, 5, 7, 11])

# 모델 생성 및 학습
model = LinearRegression()
model.fit(X, y)

# R^2 값 계산
r_square = model.score(X, y)
print(f"R^2 value: {r_square}")

# 원본 데이터 시각화
plt.scatter(X, y, color='blue', label='Data')

# 회귀선 시각화
X_range = np.linspace(1, 6, 100).reshape(-1, 1)
y_range = model.predict(X_range)
plt.plot(X_range, y_range, color='red', label='Linear Regression')

# 그래프 꾸미기
plt.title('Linear Regression Example')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 가상의 데이터 생성 (노이즈 추가)
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 3 + 4 * X + np.random.randn(100, 1) * 10  # 많은 노이즈 추가

# 모델 생성 및 학습
model = LinearRegression()
model.fit(X, y)

# R^2 값 계산
r_square = model.score(X, y)
print(f"R^2 value: {r_square}")

# 데이터 시각화
plt.scatter(X, y, color='blue', label='Data')

# 회귀선 시각화
X_range = np.linspace(0, 2, 100).reshape(-1, 1)
y_range = model.predict(X_range)
plt.plot(X_range, y_range, color='red', label='Linear Regression')

# 그래프 꾸미기
plt.title(f'Linear Regression Example (R^2 = {r_square:.2f})')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 설정
num_plots = 10
fig, axes = plt.subplots(2, 5, figsize=(20, 8))
axes = axes.flatten()

# 각 R^2 값에 대한 데이터 생성 및 시각화
for i, r_square_target in enumerate(np.linspace(0.1, 1.0, num_plots)):
    # 데이터 생성
    np.random.seed(i)
    X = np.random.rand(100, 1) * 10
    noise = np.random.randn(100, 1) * (1 - r_square_target) * 10
    y = 2 + 3 * X + noise

    # 모델 학습
    model = LinearRegression()
    model.fit(X, y)

    # R^2 값 계산
    r_square = model.score(X, y)

    # 데이터 시각화
    axes[i].scatter(X, y, color='blue', label='Data')
    X_range = np.linspace(0, 10, 100).reshape(-1, 1)
    y_range = model.predict(X_range)
    axes[i].plot(X_range, y_range, color='red', label='Linear Regression')
    axes[i].set_title(f'R^2 = {r_square:.2f}')
    axes[i].set_xlabel('X')
    axes[i].set_ylabel('y')
    axes[i].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 설정
r_square_targets = np.linspace(0.1, 1.0, 10)
num_points = 100
fig, axes = plt.subplots(2, 5, figsize=(20, 8))
axes = axes.flatten()

# 각 R^2 값에 대한 데이터 생성 및 시각화
for i, r_square_target in enumerate(r_square_targets):
    # 데이터 생성
    np.random.seed(i)
    X = np.random.rand(num_points, 1) * 10
    # 선형 관계 생성
    y_true = 2 + 3 * X
    # 목표 R^2 값을 맞추기 위해 노이즈 추가
    noise_variance = np.var(y_true) * (1 - r_square_target) / r_square_target
    noise = np.random.randn(num_points, 1) * np.sqrt(noise_variance)
    y = y_true + noise

    # 모델 학습
    model = LinearRegression()
    model.fit(X, y)

    # R^2 값 계산
    r_square = model.score(X, y)

    # 데이터 시각화
    axes[i].scatter(X, y, color='blue', label='Data')
    X_range = np.linspace(0, 10, 100).reshape(-1, 1)
    y_range = model.predict(X_range)
    axes[i].plot(X_range, y_range, color='red', label='Linear Regression')
    axes[i].set_title(f'R^2 = {r_square:.2f}')
    axes[i].set_xlabel('X')
    axes[i].set_ylabel('y')
    axes[i].legend()

plt.tight_layout()
plt.show()
```
